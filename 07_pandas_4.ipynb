```python
import warnings
warnings.filterwarnings('ignore')
from IPython.display import Image
import numpy as np
import pandas as pd
```

데이터에 누락값이 없으면 좋겠지만 누락값은 대부분의 데이터에 존재하고 깔끔한 데이터를 위해 누락값을 처리하는 것은  
매우 중요하다. 누락값은 NaN으로 표기되며 누락값을 파이썬에서 사용하려면 넘파이 라이브러리가 필요하다.

```python
# 누락값을 사용하기 위해 넘파이 누락값을 불러온다.
from numpy import NaN, NAN, nan
```

```python
# 누락값은 0, ''과 같은 값과는 다른 개념으로 데이터 자체가 없다는 것을 의미한다. => NULL => '같다'라는 개념이 존재 X
print(NaN == True)
print(NaN == False)
print(NaN == 0)
print(NaN == '')
# 누락값은 값 자체가 없기 때문에 자기 자신과 비교해도 False가 출력된다.
print(NaN == NaN)
```

```python
# 판다스에는 누락값을 확인하는 isnull(), isna() 함수가 있고 누락값이면 True, 아니면 False를 리턴한다.
print(pd.isnull(NaN))
print(pd.isna(NaN))
print('-' * 30)
print(pd.isnull(0))
```

```python
# Cases_* => 발병자 수, Deaths_* => 사망자 수

ebola = pd.read_csv('data/country_timeseries.csv')
ebola.head()
```

```python
# 넘파이의 count_nonzero() 함수는 배열에서 0이나 False가 아닌 데이터의 개수를 센다.
print(np.count_nonzero([False,0]))
print(np.count_nonzero([False,0,True,1,NaN]))
# isnull() 함수는 누락값이면 True, 누락값이 아니면 False를 리턴한다.
# isnull() 로 NaN 값이 전부 True가 되고 나머지는 False가 되니까 NaN값 개수가 나온다.
print(np.count_nonzero(ebola.isnull()))

print(np.count_nonzero(ebola['Cases_Guinea'].isnull())) # case_Guinea 열의 NaN 개수를 센다.

print(ebola['Cases_Guinea'].count()) # NaN은 count 하지 않으니까 NaN이 아닌 값
```

```python
# count() 함수로 누락값이 아닌 데이터의 개수를 알 수 있다.
print(ebola.count())
```

```python
print(ebola.shape)
print('ebola 데이터는 {}행 {}열'.format(ebola.shape[0],ebola.shape[1]))
# 전체 데이터의 개수에서 누락값이 아닌 데이터의 개수를 빼면 누락값의 개수를 알 수 있다.
num_row = ebola.shape[0]
num_missing = num_row - ebola.count()
num_missing
```

```python
# value_counts() 함수는 지정한 열의 빈도수를 얻어온다. => NaN은 무시하고 개수를 센다.
print(type(ebola['Cases_Guinea'])) # 2개 이상의 열은 리스트 형태로 열을 지정해서 얻어와야 한다.
# 열이 하나만 필요하면 밑에 처럼 해도 된다.
print(type(ebola.Cases_Guinea)) # 데이터프레임에 '.'을 찍어서 특정열의 데이터를 얻어 올 수 있다. => 시리즈
print(ebola['Cases_Guinea'].value_counts())
print(ebola.Cases_Guinea.value_counts())
```

```python
# value_counts() 함수는 dropna 옵션의 기본값이 True이고 NaN값을 무시한다. NaN도 취급하게 하려면 dropna 옵션값을 False로 지정
print(ebola.Cases_Guinea.value_counts(dropna=False))
```

```python
# 기초 통계 계산 함수 skipna 옵션의 기본값이 True 이고 NaN 값을 무시한다. NaN도 취급하게 하려면 skipna 옵션을 False로 지정
print(ebola.Cases_Guinea.sum(skipna=False))
print(ebola.Cases_Guinea.sum())
print(ebola.Cases_Guinea.mean(skipna=False)) # NaN이 포함되는 순간 계산이 안돼서 이거 XX
print(ebola.Cases_Guinea.mean())
# print(ebola.Cases_Guinea.count()) # count() 함수는 skipna 옵션을 사용할 수 없다.
```

누락값 처리하기

```python
ebola_s = ebola.iloc[0:10, 0:5]
ebola_s
```

```python
# fillna() 함수의 인수로 특정 데이터를 지정하면 누락값을 지정된 값으로 변경한다.
ebola_s.fillna(0)
```

```python
# fillna() 함수의 method 옵션을 'ffill'로 지정하면 데이터를 지정하면 누락값이 누락값 이전 값으로 변경된다.
# 이전 데이터가 없을 시 에는 변경되지 않는다.
ebola_s.fillna(method='ffill')
```

```python
# fillna() 함수의 method 옵션을 'bfill'로 지정하면 데이터를 지정하면 누락값이 누락값 다음 값으로 변경된다.
# 다음 데이터가 없을 시 에는 변경되지 않는다.
ebola_s.fillna(method='bfill')
```

```python
# NaN이 앞뒤 값의 평균값으로 변경된다. interpolate()
ebola_s.interpolate()
```

```python
# dropna() 함수로 누락값이 포함된 행을 삭제할 수 있다.
ebola_dropna = ebola.dropna()
ebola_dropna
```
